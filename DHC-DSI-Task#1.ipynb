### Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

----

### Data

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

data = pd.read_csv(url, names=names)
data.head()
```

---

```python
data.shape
```

```python
data.columns
```

-----

```python
sns.countplot(x= 'species', data = data, )
plt.show()
```

All categories present in the target column are equally distributed

----

```python
sns.scatterplot(x = 'sepal_length', y = 'sepal_width', hue = 'species', data = data, )
plt.show()
```

```python
sns.scatterplot(x = 'petal_length', y = 'petal_width', hue = 'species', data = data, )
plt.show()
```

A visible observation here is that iris-setosa is very distinct having shorter sepal lengths and greator width as compared to other flowers, however, verscolor and virginica are mostly spread out in between  

There is a strong, positive linear correlation between petal length and petal width. More importantly, the species separate beautifully. Setosa is completely isolated in the bottom left, and the boundary between versicolor and virginica is much sharper than in the sepal data.

----

```python
fig, axes = plt.subplots(2, 2, figsize=(10,10))

axes[0,0].set_title("Sepal Length")
axes[0,0].hist(data['sepal_length'], bins=15)

axes[0,1].set_title("Sepal Width")
axes[0,1].hist(data['sepal_width'], bins=15);

axes[1,0].set_title("Petal Length")
axes[1,0].hist(data['petal_length'], bins=15);

axes[1,1].set_title("Petal Width")
axes[1,1].hist(data['petal_width'], bins=15);
```

The sepal features (top row) approximate a normal distribution. However, the petal features (bottom row) are highly bimodal (two distinct peaks).

-----

```python
def graph(y):
    sns.boxplot(x="species", y=y, data=data)

plt.figure(figsize=(10,10))
    
plt.subplot(221)
graph('sepal_length')

plt.subplot(222)
graph('sepal_width')

plt.subplot(223)
graph('petal_length')

plt.subplot(224)
graph('petal_width')
```

This is the most informative view for comparing descriptive statistics across classes. It clearly shows the median values, interquartile ranges, and explicitly flags outliers (the hollow circles, such as the low petal length in versicolor or the high sepal width in virginica). It confirms that petal length and width offer the cleanest separation between classes.
